# Langkah (Step) 1

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

# Langkah (Step) 2:

```python
# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://img.okezone.com/content/2022/09/02/612/2659690/rekomendasi-400-nama-kucing-yang-unik-lucu-dan-tidak-biasa-ada-mirip-nama-makanan-1MFIjBlxVB.jpeg"
       ]
# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)                          #read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) #resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)  #convert color to RGB
  final_frame = cv.hconcat((image, image_2))      #concatenate image
  cv2_imshow(final_frame)                         #show image
  print('\n')
```

# Langkah (Step) 3
melihat ukuran FIle image

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ",tinggi," x ",lebar)
cv2_imshow(image_2)
```

# Langkah (Step) 4
mengakses pixel dengan memberi garis diagonal menyilang

```python
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# membuat garis horizontal ditengah image
for y in range (lebar):
  image_3[int((tinggi)/2),y] = [255,255,255]

final_frame= cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

# 4. Perhatikan potongan kode progam berikut:membuat garis horizontal ditengah image for y in range (lebar):  image_3[int((tinggi)/2),y] = [255,255,255] Apakah kegunaan kode [255,255,255] ? Jelaskan!
Kode [255, 255, 255] digunakan untuk mengatur warna garis horizontal yang digambar di tengah gambar menjadi putih. Dalam format RGB, nilai [255, 255, 255] merepresentasikan warna putih, di mana setiap komponen warna (merah, hijau, biru) memiliki nilai maksimal 255. Dengan menggunakan nilai ini, garis horizontal putih ditambahkan ke gambar pada baris tengah.

# 5. Jelaskan keterkaitan antara pixel dan juga resolusi gambar yang tinggi ataupun rendah!
Pixel adalah unit terkecil dalam gambar digital, dan resolusi gambar menentukan jumlah pixel dalam gambar, biasanya dinyatakan sebagai lebar x tinggi. Resolusi tinggi berarti lebih banyak pixel, menghasilkan gambar yang lebih tajam dan detail. Sebaliknya, resolusi rendah berarti lebih sedikit pixel, sehingga gambar tampak kurang jelas dan detailnya lebih kasar. Dengan kata lain, semakin tinggi resolusi, semakin baik kualitas gambar yang dapat ditampilkan.

2. Buat garis vertikal dan garis menyilang diagonal pada image keluaran

```python
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Membuat garis horizontal ditengah image
for y in range (tinggi):
  image_3[y,lebar//2] = [255,255,255]

final_frame= cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Membuat garis diagonal menyilang
for i in range(min(tinggi, lebar)):
    image_3[i, i] = [255, 255, 255]  # Garis diagonal dari kiri atas ke kanan bawah
    image_3[i, lebar - i - 1] = [255, 255, 255]  # Garis diagonal dari kanan atas ke kiri bawah

final_frame= cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

3. Buat garis horisontal berwarna putih dibagian tengah gambar dengan panjang
tertentu

```python
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Panjang garis horizontal yang diinginkan
panjang_garis = 200

# Menghitung posisi awal dan akhir garis horizontal di tengah gambar
x_awal = max((lebar // 2) - (panjang_garis // 2), 0)
x_akhir = min((lebar // 2) + (panjang_garis // 2), lebar)

# Gambar garis horizontal putih
for y in range(x_awal, x_akhir):
    image_3[tinggi // 2, y] = [255, 255, 255]

final_frame= cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

4. Buat kotak menggunakan kumpulan pixel warna putih di sembarang tempat dalam
gambar

```python
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Menentukan posisi (x, y) untuk sudut kiri atas kotak dan ukuran (lebar dan tinggi) kotak
x_start = 100  # Posisi horizontal sudut kiri atas
y_start = 50   # Posisi vertikal sudut kiri atas
lebar_kotak = 150  # Lebar kotak
tinggi_kotak = 100  # Tinggi kotak

# Menghitung posisi sudut kanan bawah kotak
x_end = min(x_start + lebar_kotak, lebar)
y_end = min(y_start + tinggi_kotak, tinggi)

# Gambar kotak putih
image_3[y_start:y_end, x_start:x_end] = [255, 255, 255]

final_frame= cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
